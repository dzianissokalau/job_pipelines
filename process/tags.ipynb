```python
import requests
from bs4 import BeautifulSoup
import shutil
import pandas as pd
import time
```

```python
link = 'https://stackoverflow.com/tags?page={}&tab=popular'
```

```python
tags = []

for page in range(1,11):
    response = requests.get(link.format(str(page)))
    time.sleep(5)
    
    response_html = BeautifulSoup(response.content, 'html.parser')
    results = response_html.find_all('a', class_='post-tag')
    
    for result in results:
        tags.append(result.text)
```
